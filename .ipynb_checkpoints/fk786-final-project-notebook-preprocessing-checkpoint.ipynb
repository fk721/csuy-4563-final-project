# Unsupervised Analysis

## Article Length and Readability Score
Perhaps there is some sort of correlation between the number of words in an article, along with its text readability? The assumption that is being tested is that fake news is easier to read, and has less words in order to appeal to a large audience.
For the *text readability score* I am using the [Flesch Reading Ease Readability Formula](https://readabilityformulas.com/flesch-reading-ease-readability-formula.php) from [textstat](https://pypi.org/project/textstat/).

```python
def extract_size_and_score(dataset):
    word_count_lst = []
    text_score_lst = []
    for index, row in dataset.iterrows():
        text = str(row['text'])
        word_count = ts.lexicon_count(text) 
        reading_score = ts.flesch_reading_ease(text)
        word_count_lst.append(word_count)
        text_score_lst.append(reading_score)
    return word_count_lst, text_score_lst
```

```python
def get_mean_and_sd(lst):
    mean = sum(lst) / len(lst)
    variance = sum([((x - mean) ** 2) for x in lst]) / len(lst) 
    sd = variance ** 0.5
    return mean, sd
```

```python
fake_news_word_counts, fake_news_text_scores = extract_size_and_score(fake_news_dataset)
real_news_word_counts, real_news_text_scores = extract_size_and_score(real_news_dataset)
```

```python
avg_words_fake, sd_words_fake = get_mean_and_sd(fake_news_word_counts)
avg_score_fake, sd_score_fake = get_mean_and_sd(fake_news_text_scores)
avg_words_real, sd_words_real = get_mean_and_sd(real_news_word_counts)
avg_score_real, sd_score_real = get_mean_and_sd(real_news_text_scores)
```

```python
print("Fake news averages {:.2f}".format(avg_words_fake),"words " \
      "( with a standard deviation of {:.2f}".format(sd_words_fake)," words ) with a " \
      "readability score of {:.2f}".format(avg_score_fake),"(with a " \
      "standard deviation of {:.2f}".format(sd_score_fake),"points)")
print("")
print("Real news averages {:.2f}".format(avg_words_real),"words " \
      "( with a standard deviation of {:.2f}".format(sd_words_real)," words ) with a " \
      "readability score of {:.2f}".format(avg_score_real),"(with a " \
      "standard deviation of {:.2f}".format(sd_score_real),"points)")
```

```python
plt.title("Fake News")
plt.xlabel("Number of Words")
plt.ylabel("Readability Score")
plt.plot(fake_news_word_counts, fake_news_text_scores, 'r+')
plt.xlim([0, 1000])
plt.ylim([0, 100])
plt.show()
```

```python
plt.title("Real News")
plt.xlabel("Number of Words")
plt.ylabel("Readability Score")
plt.plot(real_news_word_counts, real_news_text_scores, 'b+')
plt.xlim([0, 1000])
plt.ylim([0, 100])
plt.show()
```

Fake news, since it had a higher score than real news, was easier to read and understand (just like I predicted). However, fake news was a bit more verbose than real news, which was a surprise. Based on the plots, there really isn't some sort of hidden correlation between the two. Both plots looked idential, just two big messes.

# Supervised Analysis: Go With Word Frequencies

# Include $y$ and Combine the Datasets

```python
fake_news_dataset['target'] = 0
real_news_dataset['target'] = 1
df = pd.concat([fake_news_dataset, real_news_dataset])
df = df.dropna()
```

# Cleaning the Dataset
Having the `subject` and `date` did not seem like helpful features for this dataset, so they were removed. In addition, `title` was removed too, since `text` was where the bulk of the information was.

```python
df = df[['text', 'target']]
df
```

# Cleaning Text
Filter out meaningless data by removing *stopwords*.
`cleanup_text` takes a while, be warned.

```python
def cleanup_text(text):
    to_remove = nltk.corpus.stopwords.words('english')
    return " ".join([word for word in nltk.word_tokenize(text.lower()) if word not in to_remove and word.isalnum()])
```

```python
df['text'] = df['text'].apply(cleanup_text)
df['text'].replace('', np.nan, inplace=True)
df.dropna(inplace = True)
```

```python
df
```

# Logisitc Regression

```python
def logistic_regression(max_words):
    vectorizer = CountVectorizer(max_features = max_words)
    vectorizer.fit(df['text'])
    bag_of_words = vectorizer.transform(df['text'])
    sum_words = bag_of_words.sum(axis = 0)
    most_freq_words = [(word, sum_words[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
    most_freq_words =sorted(most_freq_words, key = lambda x: x[1], reverse=True)
    X = bag_of_words.toarray()
    Y = df['target'].to_numpy()
    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, shuffle = True)
    logreg = LogisticRegression()
    logreg.fit(x_train, y_train)
    training_acc = logreg.score(x_train, y_train)
    testing_acc = logreg.score(x_test, y_test)
    return training_acc, testing_acc, most_freq_words
```

# Tweaking the Hyperparameter `max_words`

```python
train_acc_lst = []
test_acc_lst = []
max_words = 15
_, _, freq_words_lst = logistic_regression(max_words)
num_words_lst = [i for i in range(1, max_words + 1)]
for num_words in num_words_lst:
    if (num_words % 5 == 0):
        print("Testing",num_words,"possible_features...")
    train_acc, test_acc, _ = logistic_regression(num_words)
    train_acc_lst.append(train_acc)
    test_acc_lst.append(test_acc)
print("---------------------")
print("Frequent Words Ranks:")
for i in range(len(freq_words_lst)):
    print(i + 1,"\t",freq_words_lst[i])
```

```python
plt.xlabel("Number of Features/Words")
plt.ylabel("Accuracy")
plt.plot(num_words_lst, train_acc_lst)
plt.plot(num_words_lst, test_acc_lst)
plt.legend(['Training Accuracy', 'Testing Accuracy'], loc='lower right')
plt.show()
```

# The Data Is Bad?
Notice the crazy increase in accuracy from around 80% to 99% once the 10th feature was included (the word *reuters*). Could this single word predict whether or not an article was fake or real?

# Checking Accuracy Based the Sole Feature `reuters`

```python
def predict_based_on_retuers():
    X = df['text'].to_numpy()
    Y = df['target'].to_numpy()
    has_reuters = []
    for text in X:
        has_reuters.append(1 if 'reuters' in text else 0)
    has_reuters_np = np.array(has_reuters)
    return np.sum(has_reuters_np == Y) / has_reuters_np.shape[0]
```

```python
acc_reuters = predict_based_on_retuers()
print("Based solely on the feature 'reuters', the accuracy is {:.2%}".format(acc_reuters))
```

# Cleaning Text (Again)
It looks like we can accurately predict whether an article is *fake* or *real* on whether or not it contains the word *reuters*. Upon a closer look, nearly all of the real news articles from the dataset are taken from *reuters* and have the word *reuters* listed as a source.
I strongly doubt we can predict whether an article is real or fake in the real world given this condition (since there are other news outlets out there). 
In order to make this assignment more fun, when cleaning the article text, I am removing the word *reuters* from the text.

```python
def cleanup_text_again(text):
    to_remove = nltk.corpus.stopwords.words('english')
    return " ".join([word for word in nltk.word_tokenize(text.lower()) if word not in to_remove and word.isalnum() and word != 'reuters'])
```

```python
df['text'] = df['text'].apply(cleanup_text_again)
df['text'].replace('', np.nan, inplace=True)
df.dropna(inplace = True)
```

# How Many Features?
How many features need to be used? Clearly, as we increase the number of features/most frequent words, both the training accuracy and testing accuracy increase. After some toying around with `CountVectorizer`, I noticed that there were over 100,000 unique words spread across the 40,000 plus articles that could potentially be used as features. Can we use all 100,000 unique words as features? Sure. Is it possible? Probably not on most machines. You would need at least 30+ GB of RAM to accomplish such a feat. 
So, how many features are enough? I guess it really depends on how satisifed one is with their testing accuracy. 
I decided to use 15 features (the top 15 most frequent words) since 15 was just a nice, round number, and it gave an okay testing accuracy (a little over 80%).

```python
num_features = 15
train_acc, test_acc, freq_words_lst = logistic_regression(num_features)
print("Frequent Words Ranks:")
for i in range(len(freq_words_lst)):
    print(i + 1,"\t",freq_words_lst[i])
print("---------------------")
print("Using",num_features," total features:")
print("\tThe training accuracy: {:.2%}".format(train_acc))
print("\tThe testing  accuracy: {:.2%}".format(test_acc))
```

Saving the preproccesed data into its own CSV file so users could just skip to this part and avoid the timely preprocessing steps described above.
